```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a DataFrame with 2000 rows and columns "Die" and "Vmin"
data = {
    'Die': range(1, 2001),
    'Vmin': np.random.normal(loc=0.7, scale=0.05, size=2000)  # Mean is 0.7, standard deviation is 0.05
}

df = pd.DataFrame(data)

# Sort the DataFrame by "Vmin" to plot the CDF
df_sorted = df.sort_values('Vmin')

# Calculate the CDF
cumulative_probs = np.arange(len(df_sorted)) / len(df_sorted)

# Plot the CDF
plt.figure(figsize=(10, 6))
plt.plot(df_sorted['Vmin'], cumulative_probs, marker='.', linestyle='none')
plt.title('Cumulative Distribution Function of Vmin')
plt.xlabel('Vmin')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.show()

# Calculate the percentiles (10%, 50%, and 90%)
percentile_10 = np.percentile(df['Vmin'], 10)
percentile_50 = np.percentile(df['Vmin'], 50)
percentile_90 = np.percentile(df['Vmin'], 90)

# Output the values as a table
percentiles_table = pd.DataFrame({
    'Percentile': [10, 50, 90],
    'Value': [percentile_10, percentile_50, percentile_90]
})

print(percentiles_table)
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a DataFrame with 2000 rows and columns "Die" and "Vmin"
data = {
    'Die': range(1, 2001),
    'Vmin': np.random.normal(loc=0.7, scale=0.05, size=2000)  # Mean is 0.7, standard deviation is 0.05
}

df = pd.DataFrame(data)

# Sort the DataFrame by "Vmin" to plot the CDF
df_sorted = df.sort_values('Vmin')

# Calculate the CDF
cumulative_probs = np.arange(len(df_sorted)) / len(df_sorted)

# Plot the CDF
plt.figure(figsize=(10, 6))
plt.plot(df_sorted['Vmin'], cumulative_probs, marker='.', linestyle='none')
plt.title('Cumulative Distribution Function of Vmin')
plt.xlabel('Vmin')
plt.ylabel('Cumulative Probability')
plt.grid(True)

# Calculate the percentiles (10%, 50%, and 90%)
percentile_10 = np.percentile(df['Vmin'], 10)
percentile_50 = np.percentile(df['Vmin'], 50)
percentile_90 = np.percentile(df['Vmin'], 90)

# Add horizontal lines at the percentiles and label them
plt.axhline(y=0.1, color='r', linestyle='--', label='10th Percentile')
plt.axhline(y=0.5, color='g', linestyle='-', label='50th Percentile (Median)')
plt.axhline(y=0.9, color='b', linestyle='--', label='90th Percentile')

# Add labels and legend
plt.text(percentile_10, 0.1 + 0.02, '10th Percentile', rotation=90, va='bottom')
plt.text(percentile_50, 0.5 + 0.02, '50th Percentile', rotation=90, va='bottom')
plt.text(percentile_90, 0.9 + 0.02, '90th Percentile', rotation=90, va='bottom')
plt.legend()

# Show the plot
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import plotly.express as px
from plotly import graph_objects as go

# Create a DataFrame with 2000 rows and columns "Die" and "Vmin"
data = {
    'Die': range(1, 2001),
    'Vmin': np.random.normal(loc=0.7, scale=0.05, size=2000)  # Mean is 0.7, standard deviation is 0.05
}

df = pd.DataFrame(data)

# Sort the DataFrame by "Vmin" to plot the CDF
df_sorted = df.sort_values('Vmin')

# Calculate the CDF
cumulative_probs = np.arange(len(df_sorted)) / len(df_sorted)

# Create a Plotly figure
fig = px.scatter(x=df_sorted['Vmin'], y=cumulative_probs, title='Cumulative Distribution Function of Vmin')

# Add horizontal lines at the percentiles (10%, 50%, and 90%)
percentile_10 = np.percentile(df['Vmin'], 10)
percentile_50 = np.percentile(df['Vmin'], 50)
percentile_90 = np.percentile(df['Vmin'], 90)

fig.add_shape(type='line', x0=percentile_10, y0=0, x1=percentile_10, y1=0.1, line=dict(color='red'))
fig.add_shape(type='line', x0=percentile_50, y0=0, x1=percentile_50, y1=0.5, line=dict(color='green'))
fig.add_shape(type='line', x0=percentile_90, y0=0, x1=percentile_90, y1=0.9, line=dict(color='blue'))

# Add labels for the percentiles
fig.add_annotation(x=percentile_10, y=0.1, text="10th Percentile", showarrow=True, arrowhead=2)
fig.add_annotation(x=percentile_50, y=0.5, text="50th Percentile", showarrow=True, arrowhead=2)
fig.add_annotation(x=percentile_90, y=0.9, text="90th Percentile", showarrow=True, arrowhead=2)

# Show the plot
fig.show()
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import plotly.express as px
from plotly import graph_objects as go

# Create a DataFrame with 2000 rows and columns "Die" and "Vmin"
data = {
    'Die': range(1, 2001),
    'Vmin': np.random.normal(loc=0.7, scale=0.05, size=2000)  # Mean is 0.7, standard deviation is 0.05
}

df = pd.DataFrame(data)

# Sort the DataFrame by "Vmin" to plot the CDF
df_sorted = df.sort_values('Vmin')

# Calculate the CDF
cumulative_probs = np.arange(len(df_sorted)) / len(df_sorted)

# Create a Plotly figure with white background
fig = px.scatter(
    x=df_sorted['Vmin'], 
    y=cumulative_probs, 
    title='Cumulative Distribution Function of Vmin'
)

# Set white background with light grey grid
fig.update_layout(
    plot_bgcolor='white',
    paper_bgcolor='white',
    xaxis=dict(
        showgrid=True,
        gridcolor='lightgrey',
        gridwidth=0.5
    ),
    yaxis=dict(
        showgrid=True,
        gridcolor='lightgrey',
        gridwidth=0.5
    )
)

# Calculate the percentiles (10%, 50%, and 90%)
percentile_10 = np.percentile(df['Vmin'], 10)
percentile_50 = np.percentile(df['Vmin'], 50)
percentile_90 = np.percentile(df['Vmin'], 90)

# Add horizontal lines at the probability levels (0.1, 0.5, 0.9)
fig.add_shape(type='line', x0=min(df['Vmin']), y0=0.1, x1=percentile_10, y1=0.1, 
              line=dict(color='red', width=1.5))
fig.add_shape(type='line', x0=min(df['Vmin']), y0=0.5, x1=percentile_50, y1=0.5, 
              line=dict(color='green', width=1.5))
fig.add_shape(type='line', x0=min(df['Vmin']), y0=0.9, x1=percentile_90, y1=0.9, 
              line=dict(color='blue', width=1.5))

# Add vertical lines at the percentiles
fig.add_shape(type='line', x0=percentile_10, y0=0, x1=percentile_10, y1=0.1, 
              line=dict(color='red', width=1.5))
fig.add_shape(type='line', x0=percentile_50, y0=0, x1=percentile_50, y1=0.5, 
              line=dict(color='green', width=1.5))
fig.add_shape(type='line', x0=percentile_90, y0=0, x1=percentile_90, y1=0.9, 
              line=dict(color='blue', width=1.5))

# Add labels for the percentiles
fig.add_annotation(x=percentile_10, y=0.12, text="10th Percentile", showarrow=True, arrowhead=2)
fig.add_annotation(x=percentile_50, y=0.52, text="50th Percentile", showarrow=True, arrowhead=2)
fig.add_annotation(x=percentile_90, y=0.92, text="90th Percentile", showarrow=True, arrowhead=2)

# Improve axis labels
fig.update_xaxes(title_text='Vmin')
fig.update_yaxes(title_text='Cumulative Probability')

# Show the plot
fig.show()
```
